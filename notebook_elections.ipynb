## Tour 2

```python
df_Tour2 = pd.read_csv('donnees/ResultatsElections2ndTour.txt', sep=';', encoding = 'latin1')
print(df_Tour2.shape)
df_Tour2.head()
```

```python
df_Tour2 = df_Tour2.reset_index(level=[1, 2])
colonnes = ['Libellé du département', 'Code de la circonscription', 'Inscrits', 'Abstentions',
            '% Abs/Ins', 'Votants', '% Vot/Ins', 'Blancs', '% Blancs/Ins', '% Blancs/Vot', 'Nuls',
            '% Nuls/Ins', '% Nuls/Vot', 'Exprimés', '% Exp/Ins', '% Exp/Vot']

liste = ['N°Panneau', 'Sexe', 'Nom', 'Prénom', 'Voix', '% Voix/Ins', '% Voix/Exp']
for name in ['Macron', 'Le Pen']:
    for elt in liste:
        colonnes.append(elt + ' ' + name)
df_Tour2.columns = colonnes

colonnesAGarder = ['Libellé du département', 'Code de la circonscription',
                   'Inscrits', 'Abstentions', 'Votants', 'Blancs', 'Nuls', 'Exprimés']
for col in df_Tour2.columns:
    if 'Voix ' in col:
        colonnesAGarder.append(col)

df_Tour2 = df_Tour2[colonnesAGarder]

for i in range(df_Tour2.shape[1]):
    if 'Voix ' in colonnesAGarder[i]:
        colonnesAGarder[i] = colonnesAGarder[i][5:]
df_Tour2.columns = colonnesAGarder
df_Tour2 = df_Tour2.reset_index(drop=True)
df_Tour2.head(10)
```

```python
df_Tour2['ind'] = df_Tour2.apply(lambda x: x['Libellé du département'] + str(x['Code de la circonscription']), axis=1)
df_Tour2 = df_Tour2.groupby('ind').sum()

for col in ['Macron', 'Le Pen']:
    df_Tour2[col] = (df_Tour2[col] * 100 / df_Tour2['Exprimés'])

df_Tour2.drop(['Code de la circonscription', 'Inscrits', 'Abstentions',
              'Votants', 'Blancs', 'Nuls', 'Exprimés'], axis=1, inplace=True)

df_Tour2.head(10)
```

```python
df_Tour2['Gagnant(e)'] = pd.Series('')
liste = ['Emmanuel MACRON', 'Marine LE PEN']
data = df_Tour2.values
for i in range(df_Tour2.shape[0]):
    amax = data[i, :-1].argmax()
    col = amax%data.shape[1]
    df_Tour2['Gagnant(e)'][i] = liste[col]
df_Tour2.head(10)
```
